```python
import numpy as np
import matplotlib.pyplot as plt

# a = 5.43  # lattice constant in angstrom for Si.
Vss_sigma = -8.13 # for Si
Vsp_sigma = 2.54
Vpp_sigma = 4.47
Vpp_pi = -1.12
E_s_c = -2  # -17.52 eV for C, -13.55 eV for Si
E_s_a = -2  # -17.52 eV for C, -13.55 eV for Si
E_p_c = 5.2   # -8.97 eV for C, -6.52 eV for Si
E_p_a = 5.2   # -8.97 eV for C, -6.52 eV for Si

#####
# d_1 = np.array([1, 1, 1])*a/4
# d_2 = np.array([1, -1, -1])*a/4
# d_3 = np.array([-1, 1, -1])*a/4 
# d_4 = np.array([-1, -1, 1])*a/4 
#####

#####
E_ss = Vss_sigma
E_sp = 5.88
E_xx = 1.71
E_xy = 7.51

#####


def hamiltonian(kx, ky, kz):
    # k = np.array([kx, ky, kz])
    # g_0 = np.exp(1j*np.dot(k,d_1)) + np.exp(1j*np.dot(k,d_2)) + np.exp(1j*np.dot(k,d_3)) + np.exp(1j*np.dot(k,d_4))
    # g_1 = np.exp(1j*np.dot(k,d_1)) + np.exp(1j*np.dot(k,d_2)) - np.exp(1j*np.dot(k,d_3)) - np.exp(1j*np.dot(k,d_4))
    # g_2 = np.exp(1j*np.dot(k,d_1)) - np.exp(1j*np.dot(k,d_2)) + np.exp(1j*np.dot(k,d_3)) - np.exp(1j*np.dot(k,d_4))
    # g_3 = np.exp(1j*np.dot(k,d_1)) - np.exp(1j*np.dot(k,d_2)) - np.exp(1j*np.dot(k,d_3)) + np.exp(1j*np.dot(k,d_4))
    g_0 =  np.cos(np.pi*kx/2)*np.cos(np.pi*ky/2)*np.cos(np.pi*kz/2) - 1j*np.sin(np.pi*kx/2)*np.sin(np.pi*ky/2)*np.sin(np.pi*kz/2)
    g_1 = -np.cos(np.pi*kx/2)*np.sin(np.pi*ky/2)*np.sin(np.pi*kz/2) + 1j*np.sin(np.pi*kx/2)*np.cos(np.pi*ky/2)*np.cos(np.pi*kz/2)
    g_2 = -np.sin(np.pi*kx/2)*np.cos(np.pi*ky/2)*np.sin(np.pi*kz/2) + 1j*np.cos(np.pi*kx/2)*np.sin(np.pi*ky/2)*np.cos(np.pi*kz/2)
    g_3 = -np.sin(np.pi*kx/2)*np.sin(np.pi*ky/2)*np.cos(np.pi*kz/2) + 1j*np.cos(np.pi*kx/2)*np.cos(np.pi*ky/2)*np.sin(np.pi*kz/2)
    
    hamil = np.array([[E_s_c, E_ss*g_0, 0, 0, 0, E_sp*g_1, E_sp*g_2, E_sp*g_3],
                    [E_ss*np.conjugate(g_0), E_s_a, -E_sp*np.conjugate(g_1), -E_sp*np.conjugate(g_2), -E_sp*np.conjugate(g_3), 0, 0, 0],
                    [0, -E_sp*g_1, E_p_c, 0, 0, E_xx*g_0, E_xy*g_3, E_xy*g_2],
                    [0, -E_sp*g_2, 0, E_p_c, 0, E_xx*g_3, E_xy*g_0, E_xy*g_1],
                    [0, -E_sp*g_3, 0, 0, E_p_c, E_xx*g_2, E_xy*g_1, E_xy*g_0],
                    [E_sp*np.conjugate(g_1), 0, E_xx*np.conjugate(g_0), E_xy*np.conjugate(g_3), E_xy*np.conjugate(g_2), E_p_a, 0, 0],
                    [E_sp*np.conjugate(g_2), 0, E_xy*np.conjugate(g_3), E_xx*np.conjugate(g_0), E_xy*np.conjugate(g_1), 0, E_p_a, 0],
                    [E_sp*np.conjugate(g_3), 0, E_xy*np.conjugate(g_2), E_xy*np.conjugate(g_1), E_xx*np.conjugate(g_0), 0, 0, E_p_a]], dtype=complex)
    return hamil

def eigv(mat):
    return np.linalg.eigvals(mat)

y6 = eigv(hamiltonian(0, 0, 0))
y5 = eigv(hamiltonian(0.1, 0.1, 0.1))
y4 = eigv(hamiltonian(0.2, 0.2, 0.2))
y3 = eigv(hamiltonian(0.3, 0.3, 0.3))
y2 = eigv(hamiltonian(0.4, 0.4, 0.4))
y1 = eigv(hamiltonian(0.5, 0.5, 0.5))
y7 = eigv(hamiltonian(0, 0, 0.2))
y8 = eigv(hamiltonian(0, 0, 0.4))
y9 = eigv(hamiltonian(0, 0, 0.6))
y10 = eigv(hamiltonian(0, 0, 0.8))
y11 = eigv(hamiltonian(0, 0, 1.0))
y12 = eigv(hamiltonian(0.05, 0.05, 1.0))
y13 = eigv(hamiltonian(0.10, 0.10, 1.0))
y14 = eigv(hamiltonian(0.15, 0.15, 1.0))
y15 = eigv(hamiltonian(0.20, 0.20, 1.0))
y16 = eigv(hamiltonian(0.25, 0.25, 1.0))
y17 = eigv(hamiltonian(0.2, 0.2, 0.8))
y18 = eigv(hamiltonian(0.15, 0.15, 0.6))
y19 = eigv(hamiltonian(0.1, 0.1, 0.4))
y20 = eigv(hamiltonian(0.05, 0.05, 0.2))
y21 = eigv(hamiltonian(0.0, 0.0, 0.0))

# for i in range(50):
#     kx = 0.5
#     ky = 0.5
#     kz = 0.5
xx = np.array([1, 1+0.866/5, 1+0.866*2/5, 1+0.866*3/5, 1+0.866*4/5, 1+0.866*5/5, 1+0.866+0.2, 1+0.866+0.4, 1+0.866+0.6, 1+0.866+0.8, 1+0.866+1.0, 1+0.866+1.0+0.0707, 1+0.866+1.0+0.0707*2, 1+0.866+1.0+0.0707*3, 1+0.866+1.0+0.0707*4, 1+0.866+1.0+0.0707*5, 1+0.866+1.0+0.0707*5+0.2, 1+0.866+1.0+0.0707*5+0.4, 1+0.866+1.0+0.0707*5+0.6, 1+0.866+1.0+0.0707*5+0.8, 1+0.866+1.0+0.0707*5+1.0])
energy = np.array([y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, y12, y13, y14, y15, y16, y17, y18, y19, y20, y21])
plt.rcParams['figure.dpi'] = 200
plt.figure()
plt.plot(xx, energy, 'bs', markersize=2)
plt.axis([0.9, 1+0.866+1.0+0.0707*5+1.0+0.1, -11, 2])
# plt.plot(xx, energy_10, 'rs-', markersize=8)
plt.ylabel('Total energy (Ry)', fontsize=14)
plt.xlabel("K-point",fontsize=14)
# plt.xaxis.set_tick_params(labelsize=14)
# plt.yaxis.set_tick_params(labelsize=14)
# plt.grid(True)
# ax[0].set_title("Paraelectric phase")
# plt.legend(['100 Ry', '10 Ry'])
# ax[0].set_xticks(xx)
# plt.savefig("Excercis2_2")
```
