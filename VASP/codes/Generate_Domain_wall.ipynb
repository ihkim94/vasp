```python
from pymatgen.io.vasp import Poscar
import numpy as np

"""
    Note that this code is only functioning for Ferroelectric Perovskite Oxides materials such as BaTiO3, and PbTiO3.
    
    Has to be improved in the future.
"""

file_name = "../VASP_files/POSCAR_LDA"
# file_name = "VASP_files/POSCAR_PBEsol"
poscar = Poscar.from_file(file_name)
struc = poscar.structure
# print(struc)


##### Input parameters to produce the domain wall supercell.
n_bulk = 6  ## This number has to be confirmed later, might has to be tested later for the convergence.
n_DW = 3  ## The size of the unit cell.
Selective_dynamics_Ba_atoms = True
#####


delta_z_Ti = (struc.frac_coords[1][2] - 0.5)/((n_DW+1)/2)
delta_z_O1 = np.abs((struc.frac_coords[2][2]) - 0.5)/((n_DW+1)/2)
delta_z_O3 = np.abs((struc.frac_coords[4][2]) - 1)/((n_DW+1)/2)
print(delta_z_Ti, delta_z_O1, delta_z_O3)
print(struc)

struc.make_supercell([[1, 0, 0], [0, (2*n_bulk + 2*n_DW), 0], [0, 0, 1]])
i = (2*n_bulk + 2*n_DW) + n_bulk

a = 1
for x in range(i, i+n_DW):
    struc.translate_sites(x, [0, 0, - a * delta_z_Ti], True, False)
    a = a + 1
    
for x in range(i+n_DW, i+n_DW+n_bulk):
    struc.translate_sites(x, [0, 0, - a * delta_z_Ti], True, False)

a = n_DW
for x in range(i+n_DW+n_bulk, i + 2*n_DW + n_bulk):
    struc.translate_sites(x, [0, 0, - a * delta_z_Ti], True, False)
    a = a - 1

## Oxygen_1 parts start

a = 1    
for x in range(i + 2*n_DW + 2*n_bulk, i + 3*n_DW + 2*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)
    a = a + 1    
    
for x in range(i + 3*n_DW + 2*n_bulk, i + 3*n_DW + 3*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)

a = n_DW
for x in range(i + 3*n_DW + 3*n_bulk, i + 4*n_DW + 3*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)
    a = a - 1

## Oxygen_2 parts start

a = 1    
for x in range(i + 4*n_DW + 4*n_bulk, i + 5*n_DW + 4*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)
    a = a + 1    
    
for x in range(i + 5*n_DW + 4*n_bulk, i + 5*n_DW + 5*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)

a = n_DW
for x in range(i + 5*n_DW + 5*n_bulk, i + 6*n_DW + 5*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O1], True, False)
    a = a - 1
    
## Oxygen_3 parts start

a = 1
for x in range(i + 6*n_DW + 6*n_bulk, i + 7*n_DW + 6*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O3], True, False)
    a = a + 1
    
for x in range(i + 7*n_DW + 6*n_bulk, i + 7*n_DW + 7*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O3], True, False)
    
a = n_DW
for x in range(i + 7*n_DW + 7*n_bulk, i + 8*n_DW + 7*n_bulk):
    struc.translate_sites(x, [0, 0, + a * delta_z_O3], True, False)
    a = a -1   
    


T = [True, True, True]
F = [False, False, False]
tensor_selective_dynamics = []
## Ba atoms part
if Selective_dynamics_Ba_atoms == False:
    for x in range(0, 2*n_bulk + 2*n_DW):
        tensor_selective_dynamics.append(F)
else:
    for x in range(0, n_bulk):
        tensor_selective_dynamics.append(F)
    for x in range(n_bulk, n_bulk+n_DW):
        tensor_selective_dynamics.append(T)
    for x in range(n_bulk+n_DW, 2*n_bulk+n_DW):
        tensor_selective_dynamics.append(F)
    for x in range(2*n_bulk+n_DW, 2*n_bulk+2*n_DW):
        tensor_selective_dynamics.append(T)            
##

##
# Other atoms part starts
##
for x in range(2*n_bulk + 2*n_DW, 3*n_bulk + 2*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(3*n_bulk + 2*n_DW, 3*n_bulk + 3*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(3*n_bulk + 3*n_DW, 4*n_bulk + 3*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(4*n_bulk + 3*n_DW, 4*n_bulk + 4*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(4*n_bulk + 4*n_DW, 5*n_bulk + 4*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(5*n_bulk + 4*n_DW, 5*n_bulk + 5*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(5*n_bulk + 5*n_DW, 6*n_bulk + 5*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(6*n_bulk + 5*n_DW, 6*n_bulk + 6*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(6*n_bulk + 6*n_DW, 7*n_bulk + 6*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(7*n_bulk + 6*n_DW, 7*n_bulk + 7*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(7*n_bulk + 7*n_DW, 8*n_bulk + 7*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(8*n_bulk + 7*n_DW, 8*n_bulk + 8*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(8*n_bulk + 8*n_DW, 9*n_bulk + 8*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(9*n_bulk + 8*n_DW, 9*n_bulk + 9*n_DW):
    tensor_selective_dynamics.append(T)
for x in range(9*n_bulk + 9*n_DW, 10*n_bulk + 9*n_DW):
    tensor_selective_dynamics.append(F)
for x in range(10*n_bulk + 9*n_DW, 10*n_bulk + 10*n_DW):
    tensor_selective_dynamics.append(T)


struc.add_site_property("selective_dynamics", tensor_selective_dynamics)


print(struc)
# print(len(struc.sites))
struc.to(filename='./POSCAR', fmt='poscar')
```

```python
from pymatgen.io.vasp import Poscar
import numpy as np

"""
    Note that this code is only functioning for Ferroelectric Perovskite Oxides materials such as BaTiO3, and PbTiO3.
    
    Has to be improved in the future.
"""

file_name = "../VASP_files/POSCAR_LDA"
# file_name = "VASP_files/POSCAR_PBEsol"
poscar = Poscar.from_file(file_name)
struc = poscar.structure
# print(struc)


##### 
# Input parameters to produce the domain wall supercell.
n_bulk = 4  ## This number has to be confirmed later, might has to be tested later for the convergence.
n_DW = 10  ## The size of the unit cell. For the paper.
cell_total = 3*n_bulk + 2*n_DW
Selective_dynamics_Ba_atoms = True
T = [True, True, True]
F = [False, False, False]
tensor_selective_dynamics = []
#####

#! Define the ferroelectric atomic displacements
# delta_z_Ti = (struc.frac_coords[1][2] - 0.5)
# delta_z_O1 = (struc.frac_coords[2][2] - 0.5)
# delta_z_O3 = (struc.frac_coords[4][2] - 1)
delta_z_Ti = struc.cart_coords[1][2] - struc.lattice.c/2
delta_z_O1 = struc.cart_coords[2][2] - struc.lattice.c/2
delta_z_O3 = struc.cart_coords[4][2] - struc.lattice.c
print("Atomic displacements are")
print(delta_z_Ti, delta_z_O1, delta_z_O3)
print()

#! Make cubic cell first
struc.translate_sites(1, [0, 0, -delta_z_Ti], False, False)
struc.translate_sites(2, [0, 0, -delta_z_O1], False, False)
struc.translate_sites(3, [0, 0, -delta_z_O1], False, False)
struc.translate_sites(4, [0, 0, -delta_z_O3], False, False)

#! Then make a supcer cell
struc.make_supercell([[1, 0, 0], [0, cell_total, 0], [0, 0, 1]])


for j in range(0, n_bulk):
    struc.translate_sites(j, [0, 0, 0], True, False) # Ba atoms
    struc.translate_sites(j + cell_total, [0, 0, +delta_z_Ti], False, False) # Ti atoms
    struc.translate_sites(j + 2*cell_total, [0, 0, +delta_z_O1], False, False) # O1 atoms
    struc.translate_sites(j + 3*cell_total, [0, 0, +delta_z_O1], False, False) # O2 atoms
    struc.translate_sites(j + 4*cell_total, [0, 0, +delta_z_O3], False, False) # O3 atoms    

i = 1
for j in range(n_bulk, n_bulk + n_DW):
    struc.translate_sites(j, [0, 0, 0], True, False) # Ba atoms
    struc.translate_sites(j + cell_total, [0, +delta_z_Ti * i/n_DW, +delta_z_Ti * ((n_DW - i)/n_DW)], False, False) # Ti atoms   
    struc.translate_sites(j + 2*cell_total, [0, +delta_z_O1 * i/n_DW, +delta_z_O1 * ((n_DW - i)/n_DW)], False, False) # O1 atoms
    struc.translate_sites(j + 3*cell_total, [0, +delta_z_O1 * i/n_DW, +delta_z_O1 * ((n_DW - i)/n_DW)], False, False) # O2 atoms
    struc.translate_sites(j + 4*cell_total, [0, +delta_z_O3 * i/n_DW, +delta_z_O3 * ((n_DW - i)/n_DW)], False, False) # O3 atoms
    i = i + 1
    
for j in range(n_bulk + n_DW, 2*n_bulk + n_DW):
    struc.translate_sites(j, [0, 0, 0], False, False) # Ba atoms
    struc.translate_sites(j + cell_total, [0, +delta_z_Ti, 0], False, False) # Ti atoms   
    struc.translate_sites(j + 2*cell_total, [0, +delta_z_O1, 0], False, False) # O1 atoms
    struc.translate_sites(j + 3*cell_total, [0, +delta_z_O1, 0], False, False) # O2 atoms
    struc.translate_sites(j + 4*cell_total, [0, +delta_z_O3, 0], False, False) # O3 atoms

i = 1
for j in range(2*n_bulk + n_DW, 2*n_bulk + 2*n_DW):
    struc.translate_sites(j, [0, 0, 0], True, False) # Ba atoms
    struc.translate_sites(j + cell_total, [0, +delta_z_Ti * ((n_DW - i)/n_DW), +delta_z_Ti * i/n_DW], False, False) # Ti atoms   
    struc.translate_sites(j + 2*cell_total, [0, +delta_z_O1 * ((n_DW - i)/n_DW), +delta_z_O1 * i/n_DW], False, False) # O1 atoms
    struc.translate_sites(j + 3*cell_total, [0, +delta_z_O1 * ((n_DW - i)/n_DW), +delta_z_O1 * i/n_DW], False, False) # O2 atoms
    struc.translate_sites(j + 4*cell_total, [0, +delta_z_O3 * ((n_DW - i)/n_DW), +delta_z_O3 * i/n_DW], False, False) # O3 atoms        
    i = i + 1

for j in range(2*n_bulk + 2*n_DW, 3*n_bulk + 2*n_DW):
    struc.translate_sites(j, [0, 0, 0], True, False) # Ba atoms
    struc.translate_sites(j + cell_total, [0, 0, +delta_z_Ti], False, False) # Ti atoms
    struc.translate_sites(j + 2*cell_total, [0, 0, +delta_z_O1], False, False) # O1 atoms
    struc.translate_sites(j + 3*cell_total, [0, 0, +delta_z_O1], False, False) # O2 atoms
    struc.translate_sites(j + 4*cell_total, [0, 0, +delta_z_O3], False, False) # O3 atoms    

#! Print final structure      
print(struc)

# T = [True, True, True]
# F = [False, False, False]
# tensor_selective_dynamics = []
# ## Ba atoms part
# if Selective_dynamics_Ba_atoms == True:
#     for x in range(0, 2*n_bulk + 2*n_DW):
#         tensor_selective_dynamics.append(F)
# else:
#     for x in range(0, n_bulk):
#         tensor_selective_dynamics.append(F)
#     for x in range(n_bulk, n_bulk+n_DW):
#         tensor_selective_dynamics.append(T)
#     for x in range(n_bulk+n_DW, 2*n_bulk+n_DW):
#         tensor_selective_dynamics.append(F)
#     for x in range(2*n_bulk+n_DW, 2*n_bulk+2*n_DW):
#         tensor_selective_dynamics.append(T)            
# ##

# ##
# # Other atoms part starts
# ##
# for x in range(2*n_bulk + 2*n_DW, 3*n_bulk + 2*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(3*n_bulk + 2*n_DW, 3*n_bulk + 3*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(3*n_bulk + 3*n_DW, 4*n_bulk + 3*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(4*n_bulk + 3*n_DW, 4*n_bulk + 4*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(4*n_bulk + 4*n_DW, 5*n_bulk + 4*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(5*n_bulk + 4*n_DW, 5*n_bulk + 5*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(5*n_bulk + 5*n_DW, 6*n_bulk + 5*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(6*n_bulk + 5*n_DW, 6*n_bulk + 6*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(6*n_bulk + 6*n_DW, 7*n_bulk + 6*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(7*n_bulk + 6*n_DW, 7*n_bulk + 7*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(7*n_bulk + 7*n_DW, 8*n_bulk + 7*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(8*n_bulk + 7*n_DW, 8*n_bulk + 8*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(8*n_bulk + 8*n_DW, 9*n_bulk + 8*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(9*n_bulk + 8*n_DW, 9*n_bulk + 9*n_DW):
#     tensor_selective_dynamics.append(T)
# for x in range(9*n_bulk + 9*n_DW, 10*n_bulk + 9*n_DW):
#     tensor_selective_dynamics.append(F)
# for x in range(10*n_bulk + 9*n_DW, 10*n_bulk + 10*n_DW):
#     tensor_selective_dynamics.append(T)

# struc.add_site_property("selective_dynamics", tensor_selective_dynamics)


# print(struc)
# # print(len(struc.sites))
struc.to(filename='./POSCAR', fmt='poscar')
```
