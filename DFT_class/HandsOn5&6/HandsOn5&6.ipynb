```python
def elastic_property(eta, volume, U_zero, U_plus, U_minus, U_tetra_plus, U_tetra_minus, U_tri_plus = 0, U_tri_minus = 0):
    C_11_plus_2C12 = (U_plus + U_minus - 2*U_zero)/2 * (1/volume) *(2/3) *(1/eta**2) * 15203.6
    C_11_minus_C12 = (U_tetra_plus + U_tetra_minus - 2*U_zero)/2 * (1/volume) *(1/3) *(1/eta**2) * 15203.6
    C_44 = (U_tri_plus + U_tri_minus - 2*U_zero)/2 * (1/volume) *(2) *(1/eta**2) * 15203.6
    C_11 = (1/3) * (C_11_plus_2C12 + 2*C_11_minus_C12)  # unit = GPa
    C_22 = (1/3) * (C_11_plus_2C12 - C_11_minus_C12) # unit = GPa
    bulk_modulus = (1/3) * C_11_plus_2C12
    print(C_11)
    print(C_22)
    print(C_44)
    print(bulk_modulus)
    

    
```
