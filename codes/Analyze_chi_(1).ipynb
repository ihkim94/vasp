```python
import ase.io
import yaml
import numpy as np
import pymatgen.io.vasp
import xml.etree.ElementTree as ET
import yaml
import ase.io
from ase.units import Bohr
from pymatgen.io.vasp import Vasprun
import numpy as np
import pandas as pd


def dielec(filename, i, j):
    """
    Extracting the electronic dielectric tensor

    Args:
        x (integer): number of directory
        i, j (integer): static dielectric tensor index (i,j = 1, 2, 3)

    Returns:
        array: electronic dielectric tensor
    """
    vrun = Vasprun(f'{filename}/vasprun.xml')
    die = vrun.epsilon_static    
    return np.array(die)[i-1][j-1]

die_11 = []
die_22 = []
die_33 = []

for i in range(1,16):
    die_11.append(dielec(f"{i:03d}", 1, 1))
    die_22.append(dielec(f"{i:03d}", 2, 2))
    die_33.append(dielec(f"{i:03d}", 3, 3))
    
print(die_11)
print(die_22)
print(die_33)
```

## Plotting the susceptibility components with respect to a certain atom and a certain direction of a displacement perturbation.

```python
import ase.io
import yaml
import numpy as np
import matplotlib.pyplot as plt
import pymatgen.io.vasp
import xml.etree.ElementTree as ET
import yaml
import ase.io
from ase.units import Bohr
from pymatgen.io.vasp import Vasprun
import numpy as np
import pandas as pd


def dielec(filename, i, j):
    """
    Extracting the electronic dielectric tensor

    Args:
        x (integer): number of directory
        i, j (integer): static dielectric tensor index (i,j = 1, 2, 3)

    Returns:
        array: electronic dielectric tensor
    """
    vrun = Vasprun(f'{filename}/vasprun.xml')
    die = vrun.epsilon_static    
    return np.array(die)[i-1][j-1]

die_11 = []
die_22 = []
die_33 = []

####### Specify the atom and the direction of the perturbation you are interested in.
n = 1 # 1 = Li_1, 2 = Li_2, 3 = Nb_1, 4 = Nb_2 ... which atom that is perturbed are you interested in?
m = 1 # 1 = x, 2 = y,  3 = z.
#######

for i in (n+(m-1)*20, n+10+(m-1)*20):
    die_11.append(dielec(f"3-6.raman_0.0025_Ang/{i:03d}", 1, 1))
    die_11.append(dielec(f"3-3.raman_0.005_Ang/{i:03d}", 1, 1))
    die_11.append(dielec(f"3-7.raman_0.0075_Ang/{i:03d}", 1, 1))
    die_11.append(dielec(f"3.raman_0.01_Ang/{i:03d}", 1, 1))
    die_11.append(dielec(f"3-4.raman_0.015_Ang/{i:03d}", 1, 1))
    die_11.append(dielec(f"3-5.raman_0.02_Ang/{i:03d}", 1, 1))

    die_22.append(dielec(f"3-6.raman_0.0025_Ang/{i:03d}", 2, 2))
    die_22.append(dielec(f"3-3.raman_0.005_Ang/{i:03d}", 2, 2))
    die_22.append(dielec(f"3-7.raman_0.0075_Ang/{i:03d}", 2, 2))
    die_22.append(dielec(f"3.raman_0.01_Ang/{i:03d}", 2, 2))
    die_22.append(dielec(f"3-4.raman_0.015_Ang/{i:03d}", 2, 2))
    die_22.append(dielec(f"3-5.raman_0.02_Ang/{i:03d}", 2, 2))
    
    die_33.append(dielec(f"3-6.raman_0.0025_Ang/{i:03d}", 3, 3))
    die_33.append(dielec(f"3-3.raman_0.005_Ang/{i:03d}", 3, 3))
    die_33.append(dielec(f"3-7.raman_0.0075_Ang/{i:03d}", 3, 3))
    die_33.append(dielec(f"3.raman_0.01_Ang/{i:03d}", 3, 3))
    die_33.append(dielec(f"3-4.raman_0.015_Ang/{i:03d}", 3, 3))
    die_33.append(dielec(f"3-5.raman_0.02_Ang/{i:03d}", 3, 3))

die_11.append(6.419711)
die_22.append(6.419711)
die_33.append(7.259361)

print("die_11 = ", die_11)
print("die_22 = ",die_22)
print("die_33 = ",die_33)
```
