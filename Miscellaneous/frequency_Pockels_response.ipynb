## Plotting frequency dependent Pockels response for the MURI review.

```python
xx = np.linspace(0, 30, 10000) # Frequency range in GHz.
T = np.linspace(0, 400, 1000)  # Temperature in K.
T_energy = T*8.617333262145 * 10**-5 # Temperature in meV.
r_ionic = 900 # Ionic Pockels coefficient in pm/V. Note that there is also temperature dependence.
r_piezo = 660 # Piezoelectric coefficient in pm/V. Note that there is no temperature dependence.
phonon_frequency = np.array([17.08, 78.17, 170.35, 174.94, 175.79, 236.19, 284.61, 294.48, 295.74, 471.44, 471.85, 495.09]) * 0.123983 # phonon frequency in meV (from cm-1).
r_ionic_i = np.array([795, 27, 0.1, 0.05, -0.02, 0.44, -0.01, 0.044, -0.026, -0.069, 0.024, 0.049]) # Ionic Pockels coefficient per each mode in pm/V.


def r_ionic_T(T):
    y = 0
    for i in range(len(phonon_frequency)):
        y += r_ionic_i[i] * (1/(np.exp(phonon_frequency[i]/T_energy) - 1))
    return y

def tanh(w, w_0, r):
    """
    To interpolate the Pockels coefficient as a function of frequency using the tanh function.
    

    Args:
        w (np.array): frequency range (GHz).
        w_0 (integer): when the response dies off. 10 for Ionic, 100 for Piezoelectric response.
        r (Pockels coefficient): Constant Pockels coefficient. 820 for Ionic, 760 for Piezoelectric response in tetragonal BaTiO3.

    Returns:
        integer: frequency dependent Pockels coefficient.
    """
    return (-np.tanh((w-w_0)/2) + 1) * 0.5 * r
    

r_ionic_w = tanh(xx, 500, r_ionic)
r_piezo_w = tanh(xx, 10, r_piezo)

r_total_w = r_ionic_w + r_piezo_w

plt.figure(figsize=(14, 8), dpi=100)  # Set DPI to 300
plt.plot(xx, r_total_w, linewidth=2)
plt.ylim(-10, 1750)
plt.xlabel('freqeuncy (GHz)', fontsize=30)
plt.ylabel(r'Pockels coefficient $r_{42}$ (pm/V)', fontsize=30)  # LaTeX syntax for fraction
plt.tick_params(axis='both', which='major', labelsize=24)  # Increase tick number size
# plt.title('Charge density')
# plt.legend()


plt.show()
```
